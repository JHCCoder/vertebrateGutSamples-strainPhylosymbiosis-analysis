```python
## What marker_in_n_samples threshold you want to investigate?
marker_threshold = "50" #65

## Category of samples wishing to run! options are "", "_mammallianHost", "_nonMammallianHost"
category = ""

## Set up the directory_path to work in 
working_dir = "/panfs/panfs1.ucsd.edu/panscratch/jhc103/VertMetaphlan-frmerged"
tree_dir = "/panfs/panfs1.ucsd.edu/panscratch/jhc103/VertMetaphlan-frmerged/strain-trees-output"
```

```python
## Import and adjust dataframe 
path_to_file = join(tree_dir, "optimal_marker_per_sample_setting_for_"+marker_threshold+"%_marker_threshold"+category+".csv")
df_bac_marker=pd.read_csv(path_to_file,index_col=0)
# df_bac_marker.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
for index, row in df_bac_marker.iterrows():
    
    ## Create the two bacteria name string you might want to utilize
    bacteria_frontTrimmed = index.strip("s__")
    bacteria_backTrimmed = index.strip("fna").strip(".")
    bacteria_trimmed = bacteria_frontTrimmed.strip("fna").strip(".")
    bacteria_simplified=bacteria_trimmed[0]+bacteria_trimmed[bacteria_trimmed.find("_"):len(bacteria_trimmed)]
    

#     ## marker_sample count
    markers_per_sample = row['marker_threshold']
    
    # Set up output directory then create it if not created yet 
    path_to_output_dir1 = join(tree_dir, "output_"+marker_threshold+"_"+bacteria_simplified)
    path_to_output_dir = join(path_to_output_dir1,"GAMMA")
    if not os.path.isdir(path_to_output_dir):
        os.mkdir(path_to_output_dir)
    
    print("Running RAxML GTRGAMMA on "+ index +":")
    
    path_to_aln = join(path_to_output_dir1, bacteria_backTrimmed+".StrainPhlAn3_concatenated.aln")
    output_name = bacteria_backTrimmed+".StrainPhlAn3.tre"

    # Run RAxML with GTRGAMMA
    ## example alignment file s__Bacteroides_vulgatus.StrainPhlAn3_concatenated.aln
    !/home/jhc103/miniconda3/envs/mpa3-py37/bin/raxmlHPC-PTHREADS-SSE3 -p 1989 -m GTRGAMMA -T 8 -w $path_to_output_dir -s $path_to_aln -n $output_name

    print("\n\n")
print("Done rebuilding all tree using GTRGAMMA model")
```
