# Import metadata & modify content 

```python
metadata_path='/panfs/panfs1.ucsd.edu/panscratch/jhc103/VertMetaphlan-frmerged/metadata/vert_metadata_new_empress_host_tree.txt'
```

```python
## Drop unwanted columns
df_metadata=pd.read_csv(metadata_path,sep="\t",index_col=0)
df_metadata.drop(["Unnamed: 27"], axis=1, inplace=True)
df_metadata
```

```python
## Replace string
df_metadata.replace(to_replace="Sarcorhamphus", value="Sarcoramphus", inplace=True)
df_metadata.loc["Callosciurus_prevostii","diet_category_3"]="herbivore"
```

## deal with columns you want (merge or sum!)

```python
df_metadata['SampleID3']=df_metadata.index
```

```python
## Check there is no empty cell in index
df_metadata['SampleID3'].value_counts()
```

```python
df_metadata.head()
```

```python
c_reads_per_file=df_metadata.groupby('SampleID3')['reads_per_file'].transform(lambda x: sum(x)/len(x)).drop_duplicates()

c_reads_per_file
```

```python
c_studyID=df_metadata.groupby('SampleID3')['studyID'].unique().transform(lambda x: '_'.join(str(v) for v in x))
print(c_studyID.to_string())
```

```python
c_host_phylum=df_metadata.groupby('SampleID3')['host_phylum'].unique().transform(lambda x: ''.join(x))
c_host_class=df_metadata.groupby('SampleID3')['host_class'].unique().transform(lambda x: ''.join(x))
c_host_order=df_metadata.groupby('SampleID3')['host_order'].unique().transform(lambda x: ''.join(x))
c_host_family=df_metadata.groupby('SampleID3')['host_family'].unique().transform(lambda x: ''.join(x))
c_host_genus=df_metadata.groupby('SampleID3')['host_genus'].unique().transform(lambda x: ''.join(x))
c_host_species=df_metadata.groupby('SampleID3')['host_species'].unique().transform(lambda x: ''.join(x))

print(c_host_species.to_string()) ## Could consider renaming host species 
```

```python
c_country=df_metadata.groupby('SampleID3')['country'].unique().transform(lambda x: np.sort(x)).transform(lambda x: '_'.join(x))
print(c_country.to_string())
```

```python
c_diet3=df_metadata.groupby('SampleID3')['diet_category_3'].unique().transform(lambda x: np.sort(x)).transform(lambda x: '_'.join(x))
print(c_diet3.to_string())
```

```python
c_diet12=df_metadata.groupby('SampleID3')['diet_category_12'].unique().transform(lambda x: np.sort(x)).transform(lambda x: '_'.join(x))
print(c_diet12.to_string())
```

```python
c_captive_wild=df_metadata.groupby('SampleID3')['captive_wild'].unique().transform(lambda x: np.sort(x)).transform(lambda x: '_'.join(x))
print(c_captive_wild.to_string())
```

### Joining the series together

```python
df_metadata_new = pd.concat([c_reads_per_file,c_studyID, c_captive_wild,c_diet3,c_country, c_host_phylum, c_host_class, c_host_order, c_host_family,c_host_genus, c_host_species], axis=1)
df_metadata_new
```

```python
## Save new column as index
df_metadata_new.insert(loc=0, column='SampleID', value=df_metadata_new.index) 
df_metadata_new.reset_index(drop=True, inplace=True)
```

```python
df_metadata_new.to_csv('/panfs/panfs1.ucsd.edu/panscratch/jhc103/VertMetaphlan-frmerged/metadata/vert_metadata_new_empress_host_tree_panda_fixed.txt', sep='\t', index=False)
```
